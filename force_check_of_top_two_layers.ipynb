```python
# import library
from ase.io import read
import numpy as np

# read QE output file
results_file = 'ontop_fcc111.qe.out'
results = read(results_file, format = 'espresso-out')

# obtain position
positions = results.get_positions()
print( positions )
```

```python
# set the z direction threshold to obtain the atoms of top two layers, including the adsorbate atom
z_threshold = 1.40842345e+01
indices = np.where(positions[:, 2] > z_threshold)[0]

if len(indices) == 0:
    raise ValueError("No atoms found above z_threshold.")


# obtain forces from QE output file
forces_all = results.get_forces()

# obtain the forces of set induces
forces_sel = forces_all[indices]

# calculate the value of force vectors and get the maximum one
max_force = np.max(np.linalg.norm(forces_sel, axis=1))

# print the number of selected atoms
print(f"Selected atoms: {len(indices)}")

# print the maximum force, which should be less than 0.02 eV/Å
print(f"Maximum force = {max_force:.5f} eV/Å")
```
